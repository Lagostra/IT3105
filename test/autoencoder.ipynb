```python
import tensorflow as tf
import tflowtools as tft
import numpy as np
import random
```

```python
input_size = 16
hidden_nodes = 8
output_size = input_size
learning_rate = 0.01
num_steps = 50000
minibatch_size = input_size // 2
```

```python
all_cases = np.array(tft.gen_all_one_hot_cases(input_size))
inputs = all_cases[:,0]
targets = all_cases[:,1]
```

```python
inpt = tf.placeholder('float')
x = inpt
y = x

w1 = tf.Variable(tf.random_uniform([input_size, hidden_nodes]))
w2 = tf.Variable(tf.random_normal([hidden_nodes, output_size]))

b1 = tf.Variable(tf.random_normal([hidden_nodes]))
b2 = tf.Variable(tf.random_normal([output_size]))
```

```python
x = tf.nn.relu(tf.add(tf.matmul(x, w1), b1), name='layer1')
x = tf.nn.relu(tf.add(tf.matmul(x, w2), b2), name='layer2')
```

```python
#loss = tf.reduce_mean(tf.pow(y - x, 2))
loss = tf.nn.softmax_cross_entropy_with_logits_v2(logits=x, labels=y)
train_op = tf.train.AdamOptimizer(learning_rate).minimize(loss)
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for i in range(1, num_steps + 1):
        minibatch = np.array(random.sample(list(inputs), minibatch_size))
        
        _, l = sess.run([train_op, loss], feed_dict={inpt: minibatch})
        
        if i % 50 == 0:
            print('[Step {}] Loss: {}'.format(i, l))
    
    softmax = tf.nn.softmax(x)
    argmax = tf.argmax(softmax)
    one_hot = tf.one_hot(argmax, 8)
    result = sess.run([argmax], feed_dict={inpt: inputs})
    print(result)
    print(inputs)
```
